# Time Series Forecasting 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import mean_squared_error
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv(r'data/PJME_hourly.csv')
```

```python
df = df.set_index('Datetime')
df.index = pd.to_datetime(df.index)
```

```python
df.plot(style='.',
        figsize=(15, 5),
        color=color_pal[0],
        title='PJME Energy Use in MW')
plt.show()
```

## Train/Test Split

```python
train = df.loc[df.index < '01-01-2015']
test = df.loc[df.index >= '01-01-2015']

fig, ax = plt.subplots(figsize=(15, 5))
train.plot(ax=ax, label='Training Set', title='Data Train/Test Split')
test.plot(ax=ax, label='Test Set')
ax.axvline('01-01-2015', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
plt.show()
```

```python
df.loc[(df.index > '01-01-2010') & (df.index < '01-08-2010')].plot(figsize=(15,5), title='Week of Data')
plt.show()
```

## Feature Creation

```python
def create_features(df):
    """
    Create time series features based on time series index
    """
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df
```

```python
df = create_features(df)
```

## Visualize our Feature to Target Relationship

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='hour', y='PJME_MW')
ax.set_title('MW by Hour')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='month', y='PJME_MW', palette='Blues')
ax.set_title('MW by Month')
plt.show()
```

## Create our Model

```python
train = create_features(train)
test = create_features(test)
```

```python
df.columns
```

```python
FEATURES = ['hour', 'dayofweek', 'quarter', 'month', 'year', 'dayofyear']
TARGET = 'PJME_MW'
```

```python
X_train = train[FEATURES]
y_train = train[TARGET]
X_test = test[FEATURES]
y_test = test[TARGET]
```

```python
reg = xgb.XGBRegressor(n_estimators=1000, early_stopping_rounds=50, learning_rate=0.01)
reg.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], verbose=100)
```

## Feature Importance

```python
fi = pd.DataFrame(data=reg.feature_importances_, 
             index=reg.feature_names_in_,
             columns=['importance'])
```

```python
fi
```

```python
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
#higly correlated features like month and dayofyear
```

## Forecast on Test

```python
test['prediction'] = reg.predict(X_test)
```

```python
df = df.merge(test[['prediction']], how='left', left_index=True, right_index=True)
```

```python
ax = df[['PJME_MW']].plot(figsize=(15,5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['True Data', 'Predictions'])
ax.set_title('Raw Data and Prediction')
plt.show()
```

```python
ax = df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['PJME_MW'].plot(figsize=(15,5), title='Week of Data')
df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['prediction'].plot(style='--')
plt.legend(['True Data', 'Prediction'])
plt.show()
```

```python
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
print(f'RMSE Score on Test Set: {score:0.2f}')
```

## Calculate Error
#### Look at the best and worst predicted days

```python
test['error'] = np.abs(test[TARGET] - test['prediction'])
```

```python
test['date'] = test.index.date
```

```python
test.groupby(['date'])['error'].mean().sort_values(ascending=False).head(5)
```

```python
test.groupby(['date'])['error'].mean().sort_values(ascending=False).tail(5)
```
